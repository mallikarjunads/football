```python
def malli(plot,title,xlabel,ylabel):
    plot.set_title(title,fontsize=13,weight="bold")
    plot.set_xlabel(xlabel,fontsize=13,weight="bold")
    plot.set_ylabel(ylabel,fontsize=13,weight="bold")
def manoj(length,breadth,style):
    plt.figure(figsize=(length,breadth))
    sns.set_style(style)
  
```

```python
manoj(12,13,"darkgrid")
ax=sns.histplot(fifa["Age"],bins=20,color="orange")
malli(ax,"players ages count","player","count")
for i in ax.containers:
    ax.bar_label(i)
```
